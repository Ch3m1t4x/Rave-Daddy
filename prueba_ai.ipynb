```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime

HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/122.0.0.0 Safari/537.36"
}


def scrape_ra_events(city: str):
    """
    Scrapea RA Guide para una ciudad concreta.
    Devuelve una lista de eventos con:
    - title
    - venue
    - date
    - url
    """

    formatted_city = city.lower().replace(" ", "")
    url = f"https://es.ra.co/events/es/{formatted_city}"

    response = requests.get(url, headers=HEADERS, timeout=10)

    if response.status_code != 200:
        return [404]

    soup = BeautifulSoup(response.text, "html.parser")

    events = []
    event_cards = soup.select("div.guide-event")  # selector real de RA

    for card in event_cards:
        # título
        title_tag = card.select_one("h3.title")
        title = title_tag.get_text(strip=True) if title_tag else "Unknown event"

        # venue
        venue_tag = card.select_one("span.venue")
        venue = venue_tag.get_text(strip=True) if venue_tag else "Unknown venue"

        # fecha
        date_tag = card.select_one("span.date")
        date_raw = date_tag.get_text(strip=True) if date_tag else "Unknown date"

        # URL
        link_tag = card.select_one("a")
        url = "https://ra.co" + link_tag["href"] if link_tag else ""

        events.append({
            "title": title,
            "venue": venue,
            "date": date_raw,
            "url": url
        })

    return events
```

```python
from langchain.tools import tool

```

```python
@tool(description="Scrape techno events from Resident Advisor by city name")
def get_ra_events(city: str) -> str:
    events = scrape_ra_events(city)

    if not events:
        return "No events found or RA could not be reached."

    formatted = ""
    for e in events[:5]:  # limitar para evitar walls of text
        formatted += (
            f"- {e['title']} @ {e['venue']} ({e['date']})\n  {e['url']}\n\n"
        )
    return formatted.strip()
```

```python
#para la realidad
from playwright.sync_api import sync_playwright
from bs4 import BeautifulSoup

URL = "https://es.ra.co/events/es/madrid"

def scrape_ra():
    with sync_playwright() as p:
        browser = p.chromium.launch(headless=True)
        page = browser.new_page()

        print("⏳ Cargando página...")
        page.goto(URL, timeout=60000)

        html = page.content()
        soup = BeautifulSoup(html, "html.parser")

        events_data = []

        # Cada evento está dentro de elementos <li role="article">
        events = soup.select("li[data-test='event-card']")

        for ev in events:
            title_el = ev.select_one("[data-test='event-card-title']")
            venue_el = ev.select_one("[data-test='event-card-venue']")
            date_el  = ev.select_one("[data-test='event-card-date']")
            link_el  = ev.select_one("a")

            event_info = {
                "title": title_el.get_text(strip=True) if title_el else None,
                "venue": venue_el.get_text(strip=True) if venue_el else None,
                "date": date_el.get_text(strip=True) if date_el else None,
                "url": "https://es.ra.co" + link_el["href"] if link_el else None
            }

            events_data.append(event_info)

        browser.close()
        return events_data


if __name__ == "__main__":
    events = scrape_ra()

    print("\n🎉 Eventos encontrados:")
    for e in events:
        print(f"- {e['date']} | {e['title']} @ {e['venue']}")
        print(f"  {e['url']}")
```

```python
import asyncio
from playwright.async_api import async_playwright
from bs4 import BeautifulSoup

URL = "https://es.ra.co/events/es/madrid"

async def scrape_ra_async():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        print("⏳ Cargando página...")
        await page.goto(URL, timeout=60000)

        html = await page.content()
        await browser.close()

        soup = BeautifulSoup(html, "html.parser")

        events_data = []
        events = soup.select("li[data-test='event-card']")

        for ev in events:
            title_el = ev.select_one("[data-test='event-card-title']")
            venue_el = ev.select_one("[data-test='event-card-venue']")
            date_el  = ev.select_one("[data-test='event-card-date']")
            link_el  = ev.select_one("a")

            events_data.append({
                "title": title_el.get_text(strip=True) if title_el else None,
                "venue": venue_el.get_text(strip=True) if venue_el else None,
                "date": date_el.get_text(strip=True) if date_el else None,
                "url": "https://es.ra.co" + link_el["href"] if link_el else None
            })

        return events_data


# ------------- Ejecutar en Jupyter -------------
events = await scrape_ra_async()

print("🎉 EVENTOS ENCONTRADOS:")
for e in events:
    print(f"- {e['date']} | {e['title']} @ {e['venue']}")
    print(f"  {e['url']}")
```
